```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results_sexWork_df = pd.read_csv('../../data/raw/prepped_online_sex_work.csv')
results_sexWork_df.head()
```

```python
results_sexWork_df['Total_chat_time'] = pd.to_timedelta(results_sexWork_df['Total_chat_time'])
```

```python
results_sexWork_df.head()
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

friend_count = results_sexWork_df.groupby(['Gender', 'Sexual_polarity'])['Friends'].mean()
friend_count = friend_count.unstack()
friend_count_graph = friend_count.plot(kind='bar', figsize=(15, 9), rot=0, fontsize=18, color=colours)

friend_count_graph.set_title('Average Number of Online Friends Based on Sexual Polarity and Gender', fontsize=24)
friend_count_graph.set_xlabel('Gender', fontsize=18)
friend_count_graph.set_ylabel('Average Number of Friends', fontsize=18)
friend_count_graph.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

comment_count = results_sexWork_df.groupby(['Gender', 'Sexual_polarity'])['Number_of_Comments_in_public_forum'].mean()
comment_count = comment_count.unstack()
comment_count_graph = comment_count.plot(kind='bar', figsize=(15, 9), rot=0, fontsize=18, color=colours)

comment_count_graph.set_title('Average Number of Comments Posted Based on Sexual Polarity and Gender', fontsize=24)
comment_count_graph.set_xlabel('Gender', fontsize=18)
comment_count_graph.set_ylabel('Average Number of Comments', fontsize=18)
comment_count_graph.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

picture_count = results_sexWork_df.groupby(['Gender', 'Sexual_polarity'])['Profile_pictures'].mean()
picture_count = picture_count.unstack()
picture_count_graph = picture_count.plot(kind='bar', figsize=(15, 9), rot=0, fontsize=18, color=colours)

picture_count_graph.set_title('Average Number of Profile Pictures Based on Sexual Polarity and Gender', fontsize=24)
picture_count_graph.set_xlabel('Gender', fontsize=18)
picture_count_graph.set_ylabel('Average Number of Profile Pictures', fontsize=18)
picture_count_graph.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

meetings_count = results_sexWork_df.groupby(['Gender', 'Sexual_polarity'])['Number_of_offline_meetings_attended'].mean()
meetings_count = meetings_count.unstack()
meetings_count_graph = meetings_count.plot(kind='bar', figsize=(15, 9), rot=0, fontsize=18, color=colours)

meetings_count_graph.set_title('Average Number of Offline Meetings Based on Sexual Polarity and Gender', fontsize=24)
meetings_count_graph.set_xlabel('Gender', fontsize=18)
meetings_count_graph.set_ylabel('Average Number of Offline Meetings', fontsize=18)
meetings_count_graph.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

friend_count = results_sexWork_df.groupby('Sexual_polarity')[['Friends']].mean()
friend_count_graph = friend_count.plot(kind='pie', figsize=(11, 11), rot=0, fontsize=20, subplots=True, autopct='%.2f', colors=colours)

plt.axis('equal')
plt.title('Percentage of Average Friend Count', fontsize=20)
plt.ylabel('', fontsize=14)

plt.legend(
   labels=friend_count.index,
   loc='upper left',
   ncol=1,
   markerscale=1.5,
   fontsize=15)

plt.show()
```

```python
colours = ['#75ECFD', '#FFFF66', '#95FF4F']

chat_time = results_sexWork_df.groupby(['Sexual_polarity'])[['Total_chat_time']].sum() / results_sexWork_df.groupby(['Sexual_polarity'])[['Total_chat_time']].count()
chat_time_graph = chat_time.plot(kind='bar', figsize=(11, 11), rot=0, fontsize=20, subplots=True, autopct='%.2f', colors=colours)

plt.axis('equal')
plt.title('Percentage of Average Friend Count', fontsize=20)
plt.ylabel('', fontsize=14)

plt.legend(
   labels=chat_time.index,
   loc='upper left',
   ncol=1,
   markerscale=1.5,
   fontsize=15)

plt.show()
```
