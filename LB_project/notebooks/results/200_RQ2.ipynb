```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.plotly as py
# import plotly.graph_objs as go
# from plotly.graph_objs import *
```

```python
results_redditThread_df = pd.read_csv('../../data/raw/prepped_foreveralone.csv')
results_redditThread_df.head()
```

```python
colours = ['#FFC8F2', '#BBDAFF']

sex_and_wellbeing = results_redditThread_df.groupby(['prostitution_legal','bodyweight'])[['social_fear']].apply(lambda x: x['social_fear'].value_counts()).unstack()

sex_and_wellbeing = sex_and_wellbeing.unstack()
graph_sex_and_wellbeing = sex_and_wellbeing.plot(kind='bar', figsize=(20, 13), color=colours, rot=0, fontsize=21)

graph_sex_and_wellbeing.set_xlabel('bodyweight', fontsize=21)
graph_sex_and_wellbeing.set_ylabel(' Count of Friends', fontsize=21)
graph_sex_and_wellbeing.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```
