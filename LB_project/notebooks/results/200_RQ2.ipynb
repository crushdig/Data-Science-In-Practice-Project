```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.plotly as py
# import plotly.graph_objs as go
# from plotly.graph_objs import *
```

```python
results_redditThread_df = pd.read_csv('../../data/raw/prepped_foreveralone.csv')
results_redditThread_df.head()
```

```python
colours = ['#ffff00', '#9400d3']

sex_and_wellbeing = results_redditThread_df.groupby(['bodyweight', 'prostitution_legal'])[['social_fear']].apply(lambda x: x['social_fear'].value_counts())

sex_and_wellbeing = sex_and_wellbeing.unstack()
graph_sex_and_wellbeing = sex_and_wellbeing.plot(kind='bar', figsize=(20, 13), color=colours, rot=0, fontsize=21)

graph_sex_and_wellbeing.set_xlabel('Body Type', fontsize=21)
graph_sex_and_wellbeing.set_ylabel('Social Fear Count', fontsize=21)
graph_sex_and_wellbeing.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```

```python
colours = ['#722f37', '#C5B358']

sex_and_wellbeing = results_redditThread_df.groupby(['pay_for_sex'])[['social_fear']].apply(lambda x: pd.DataFrame(x['social_fear'].value_counts()))

sex_and_wellbeing = sex_and_wellbeing.unstack()
graph_sex_and_wellbeing = sex_and_wellbeing.plot(kind='bar', figsize=(20, 13), color=colours, rot=45, fontsize=21)

graph_sex_and_wellbeing.set_xlabel('Sex payments', fontsize=21)
graph_sex_and_wellbeing.set_ylabel('Social Fear Count', fontsize=21)
graph_sex_and_wellbeing.legend(
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```
