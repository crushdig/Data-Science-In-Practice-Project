```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.plotly as py
# import plotly.graph_objs as go
# from plotly.graph_objs import *
```

```python
results_redditThread_df = pd.read_csv('../../data/raw/prepped_foreveralone.csv')
results_redditThread_df.head()
```

```python
colours = ['#CD919E', '#A52A2A', '#BDFCC9', '#6CA6CD']

sex_and_wellbeing = results_redditThread_df.groupby(['prostitution_legal'])[['bodyweight']].apply(lambda x: pd.DataFrame(x['bodyweight'].value_counts()))

sex_and_wellbeing = sex_and_wellbeing.unstack()
graph_sex_and_wellbeing = sex_and_wellbeing.plot(kind='bar', figsize=(20, 13), color=colours, rot=0, fontsize=21)

graph_sex_and_wellbeing.set_title('Number of People who would Legalise Prostitution based on Bodyweight', fontsize=24)
graph_sex_and_wellbeing.set_xlabel('Legalise Prostitution', fontsize=21)
graph_sex_and_wellbeing.set_ylabel('Number of People with Certain Weight', fontsize=21)
graph_sex_and_wellbeing.legend(
['Normal Weight', 'Obese', 'Overweight', 'Underweight'],
   loc='best',
   ncol=1,
   markerscale=1.5,
   fontsize=15)
```
