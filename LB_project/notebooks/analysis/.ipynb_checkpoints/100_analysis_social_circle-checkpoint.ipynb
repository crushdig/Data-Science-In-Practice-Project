## Question 1: How does a personâ€™s sexual behaviour affect their social circle?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prepped_sexWork_df = pd.read_csv('../../data/raw/prepped_online_sex_work.csv')
prepped_sexWork_df.head()
```

```python
prepped_sexWork_df['Total_chat_time'] = pd.to_timedelta(prepped_sexWork_df['Total_chat_time'])
```

```python
prepped_sexWork_df.info()
```

```python
#A look into the number of friends people have on average based on sexual orientation
prepped_sexWork_df.groupby('Sexual_polarity')[['Friends']].mean()
```

A look into the average number of friends people of different sexual polarity's have. From the overview of this dataset we can see that people whose preferences are Switch, have more friends on average. It's possible people with a Switch polarity have more popularity simply because they like to alternate between Dominant and Submissive and people might prefer the variety.

```python
#Delving further into the number of friends people have on average based on sexual orientation and
#considering what the factors contributing to their social circle are
prepped_sexWork_df.groupby('Sexual_polarity')[['Friends', 'Days', 'Profile_pictures', 'Number_of_Comments_in_public_forum', 'Number_of_offline_meetings_attended']].mean()
```

```python
prepped_sexWork_df.groupby(['Sexual_polarity'])[['Total_chat_time']].sum() / prepped_sexWork_df.groupby(['Sexual_polarity'])[['Total_chat_time']].count()
```

On further expansion of the dataset, it can be noted that people who highlighted their interest in the switch polarity, had less profile pictures uploaded, were less active publicly i.e made less comments in the forum, compared to those who had a preference in "dominant" and "submissive" respectively. There were less noted offline meetings for people into the Switch polarity although not the least but yet they still had more friends.

Exploring further, we looked into the chat times for each group and the Switch group also had the least time invested in conversation.

```python
#Delving further into the number of friends people have on average based on sexual orientation and
#observing the differences between male and females.
prepped_sexWork_df.groupby(['Sexual_polarity', 'Gender'])[['Friends', 'Days', 'Profile_pictures', 'Number_of_Comments_in_public_forum', 'Number_of_offline_meetings_attended']].mean()
```

## Additional Analysis

```python
avg_chat_time = prepped_sexWork_df.groupby(['Gender', 'Sexual_orientation'])[['Total_chat_time']].sum() / prepped_sexWork_df.groupby(['Gender', 'Sexual_orientation'])[['Total_chat_time']].count()
avg_chat_time
```

The amount of time spent by genders of different orientations was explored. It can be noted, looking at all the times that women in this dataset have invested more time chatting than men.
