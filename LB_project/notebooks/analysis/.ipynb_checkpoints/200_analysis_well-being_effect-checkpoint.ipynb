## Question 2 How does a person’s (perceived) wellbeing—anxiety, body type, etc.—affect sexual behaviour?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prepped_redditThread_df = pd.read_csv('../../data/raw/prepped_foreveralone.csv')
prepped_redditThread_df.head()
```

```python
prepped_redditThread_df.groupby(['pay_for_sex','bodyweight'])[['social_fear']].apply(lambda x: pd.DataFrame(x['social_fear'].value_counts())).unstack()
```

In the table above, each row contains a set of numbers which represents a count for the number of people who claim to have a social fear or claim not to have a social fear that may or may not have paid for sex. It also displays a count of people who identify with certain weight classes that may or may not have paid for sex.

```python
prepped_redditThread_df.groupby(['prostitution_legal','bodyweight'])[['social_fear']].apply(lambda x: x['social_fear'].value_counts()).unstack()
```

Much like the previous table, we're exploring the counts of users under social fear and bodyweight columns but in this case we observe how many people in this dataset would legalise or wouldn't legalise prostitution.

```python
prepped_redditThread_df.groupby(['prostitution_legal','pay_for_sex','bodyweight'])[['social_fear']].apply(lambda x: pd.DataFrame(x['social_fear'].value_counts())).unstack()
```

Here we're exploring the counts of users under social fear and bodyweight columns while looking at how many people would/wouldn't pay for sex **and** would legalise/wouldn't legalise prostitution.
